We need to convert our dependant variable into such notation, so that our ML model understands it.

```python
data['income'] = data['income'].map({'<=50K': 0, '>50K': 1}).astype(int)
```

Now lets convert the categorical data into numerical. First lets see all the unique values.

```python
for x in data.columns:
    print ("----------%s" % x)
    print (data[x].value_counts())
```

Now that I had a look at the distinct values, I see that on our data, columns workclass, occupation, native-country has NaN values, so let's remove them very quick.

```python
import numpy as np

data['workclass'] = data['workclass'].replace('?',np.nan)
data['occupation'] = data['occupation'].replace('?',np.nan)

data.dropna(how='any',inplace=True)
```

```python
data.isin(['?']).sum(axis=0)
```

Now lets convert all the categorical values to numerical.

```python
data['workclass'] = data['workclass'].map({'Private': 0, 'Self-emp-not-inc': 1,'Local-gov': 2, 'State-gov': 3, 'Self-emp-inc': 4,'Federal-gov': 5, 'Without-pay': 6, 'Never-worked': 7}).astype(int)
data['education'] = data['education'].map({'HS-grad': 0, 'Some-college': 1, 'Bachelors': 2, 'Masters': 3, 'Assoc-voc': 4, '11th': 5, 'Assoc-acdm': 6, '10th': 7,'7th-8th': 8, 'Prof-school': 9, '9th': 10, '12th': 11, 'Doctorate': 12, '5th-6th': 13,'1st-4th': 14,'Preschool': 15}).astype(int)
data['marital-status'] = data['marital-status'].map({'Married-civ-spouse': 0, 'Never-married': 1, 'Divorced': 2, 'Separated': 3, 'Widowed': 4,'Married-spouse-absent': 5, 'Married-AF-spouse': 6}).astype(int)
data['occupation'] = data['occupation'].map({ 'Prof-specialty': 1, 'Craft-repair': 2, 'Exec-managerial': 3, 'Adm-clerical': 4, 'Sales': 5,'Other-service': 6, 'Machine-op-inspct': 7,'Transport-moving': 8,'Handlers-cleaners': 9,'Farming-fishing': 10, 'Tech-support': 11, 'Protective-serv': 12, 'Priv-house-serv': 13,'Armed-Forces':14}).astype(int)
data['relationship'] = data['relationship'].map({'Husband': 0, 'Not-in-family': 1, 'Own-child': 2, 'Unmarried': 3,'Wife': 4,'Other-relative': 5}).astype(int)
data['race'] = data['race'].map({'White': 0, 'Black': 1, 'Asian-Pac-Islander': 2, 'Amer-Indian-Eskimo': 3, 'Other': 4}).astype(int)
data['sex'] = data['sex'].map({'Male':0, 'Female':1}).astype(int)
```

```python
data.head()
```

## Training our data

```python
X = pd.DataFrame(np.c_[data['age'], data['workclass'], data['education'],data['marital-status'],data['occupation'],data['relationship'],data['race'], data['sex'], data['hours-per-week']], columns = ['age','workclass','education','marital-status','occupation','relationship','race', 'sex', 'hours-per-week'])
y = data['income']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
regressor = LogisticRegression()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=None)
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
```

```python
from sklearn import metrics
metrics.accuracy_score(y_test, y_pred)
```

```python
regressor.coef_
```

```python
regressor.intercept_
```

```python
regressor.score(X_train, y_train)
```

```python
metrics.confusion_matrix(y_test, y_pred)
```

Diagonal line indicating number of times our model predicted true, non-diagonal are the number of wrong predictions

```python
from sklearn.metrics import classification_report
target_names = ['<=50k', '>=50k']
print(classification_report(y_test, y_pred, target_names=target_names))
```
