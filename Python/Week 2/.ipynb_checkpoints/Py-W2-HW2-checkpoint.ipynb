8. Capitalize the first letter, lowercase, uppercase, swapcase, title-case of all the elements of a given array. Use np.char.capitalize(), np.char.lower(), np.char.upper(), np.char.swapcase(), np.char.title().

```python
vec = np.array(["Shahin", "emil", "Ramil", "cavid"])
vec = np.char.capitalize(vec)
print(vec)
vec = np.char.lower(vec)
print(vec)
vec = np.char.upper(vec)
print(vec)
vec = np.char.swapcase(vec)
print(vec)
vec = np.char.title(vec)
print(vec)
```

9. Get the dates of yesterday, today and tomorrow using np.datetime64() and np.timedelta64().

```python
yesterday = np.datetime64('today', 'D') - np.timedelta64(1, 'D')
print(yesterday)

today = np.datetime64('today', 'D')
print(today)

tomorrow  = np.datetime64('today', 'D') + np.timedelta64(1, 'D')
print(tomorrow)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import pandas as pd
cars = pd.read_excel('cars.xlsx')
```

```python
cars
```

1. Show relationship between highway and city miles per gallon. \
    a. Use scatter in matplotlib.\
    b. Interpret how are variables correlated according to scatter.

```python
city_mpg = cars["cty"].tolist()
highway_mpg = cars["hwy"].tolist()

plt.scatter(city_mpg, highway_mpg)
plt.xlabel("city mpg")
plt.ylabel("highway mpg")
```

There is high correlation, as with the increase of city mpg, highway mpg also increases linearly.

2. Show distributions and scatters between all variables. While hue equals to type of
drive train which variables have the lowest and the highest correlation?
Find according to scatterplot.

```python
sns.pairplot(cars, hue = "drv")
```

By looking at the scatter plots, we can say that the displ x year, year x cyl, year x cty, year x hwy have the lowest correlation. displ x cyl have high correlation because as the number of cylinders increases, engine size increases also. displ x cty and displ x hwy have good correlation, not as strong though. cty x year and hwy x year has slight correlation, where cars from 2010 have better mileage, and front wheel drive cars have better mileage then 4wd and rwd. cty x hwy have the strongest correlation.
