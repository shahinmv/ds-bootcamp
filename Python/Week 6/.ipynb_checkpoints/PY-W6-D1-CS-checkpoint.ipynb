You will use the Taiwanese Bankruptcy Prediction data set. The dataset correspondsto a
classification problem on which you need to make predictions on the basis of whetherthe company
can have a risk to bankrupt considering features in the dataset. You are asked to use this data to
build Bagging Classifier and Boosting Classifiers model.

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('data.csv')
```

```python
data[data.duplicated()]
```

```python
data.isna().sum()
```

```python
data.head()
```

```python
data.columns = data.columns.str.strip()
```

```python
x_list = []

for x in data.columns[1:]:
    x_list.append(data[x])
```

```python
X = pd.DataFrame(np.c_[x_list]).T
y = data['Bankrupt?']
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=None)

from sklearn.preprocessing import RobustScaler    
st_x= RobustScaler()    
X_train= st_x.fit_transform(X_train)    
X_test= st_x.transform(X_test) 
```

## Training

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
```

### Bagging

```python
Bagging = BaggingClassifier(random_state=42)
Bagging.fit(X_train, y_train)
y_pred_bagging = Bagging.predict(X_test)
```

```python
from sklearn import metrics
metrics.accuracy_score(y_test, y_pred_bagging)
```

### Gradient Boosting

```python
GBC = GradientBoostingClassifier()
GBC.fit(X_train, y_train)
y_pred_gbc = GBC.predict(X_test)
```

```python
metrics.accuracy_score(y_test, y_pred_gbc)
```

### XGBoost

```python
XGB = XGBClassifier()
XGB.fit(X_train, y_train)
y_pred_xgb = XGB.predict(X_test)
```

```python
metrics.accuracy_score(y_test, y_pred_xgb)
```

### Cat Boost

```python
cat = CatBoostClassifier()
cat.fit(X_train, y_train)
y_pred_cat = cat.predict(X_test)
```

```python
metrics.accuracy_score(y_test, y_pred_cat)
```
