Dream Housing Finance company deals in all kinds of home loans. They have presence across all
urban, semi urban and rural areas. Customer first applies for home loan and after that company
validates the customer eligibility for loan.
Company wants to automate the loan eligibility process (real time) based on customer detail
provided while filling online application form. These details are Gender, Marital Status, Education,
Number of Dependents, Income, Loan Amount, Credit History and others. To automate this process,
they have provided a dataset to identify the customers segments that are eligible for loan amount so
that they can specifically target these customers.

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('loan_prediction.csv')
```

```python
data.head()
```

## Data preparation/cleaning

```python
data.info()
```

```python
pd.isnull(data).sum()
```

```python
data.dropna(inplace=True)
```

```python
pd.isnull(data).sum()
```

```python
data.info()
```

```python
for x in data.columns:
    print ("----------%s" % x)
    print (data[x].value_counts())
```

```python
data['Loan_Status'] = data['Loan_Status'].map({'Y': 1, 'N': 0}).astype(int)
data['Gender'] = data['Gender'].map({'Male': 1, 'Female': 0}).astype(int)
data['Education'] = data['Education'].map({'Graduate': 1, 'Not Graduate': 0}).astype(int)
data['Self_Employed'] = data['Self_Employed'].map({'Yes': 1, 'No': 0}).astype(int)
data['Property_Area'] = data['Property_Area'].map({'Semiurban': 0, 'Urban': 1, 'Rural': 2}).astype(int)
data['Married'] = data['Married'].map({'Yes': 1, 'No': 0}).astype(int)
data['Dependents'] = data['Dependents'].map({'3+': 3, '0': 0, '1': 1, '2': 2}).astype(int)
```

```python
data.head()
```

```python
data[data.duplicated()]
```

```python
from sklearn.model_selection import train_test_split

X = pd.DataFrame(np.c_[data['Gender'], data['Married'], data['Dependents'], data['Education'], data['Self_Employed'], data['ApplicantIncome'], data['CoapplicantIncome'], data['LoanAmount'], data['Loan_Amount_Term'], data['Credit_History'], data['Property_Area']])
y = data['Loan_Status']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=None)

from sklearn.preprocessing import StandardScaler    
st_x= StandardScaler()    
X_train= st_x.fit_transform(X_train)    
X_test= st_x.transform(X_test)  
```

## Training

```python
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
```

```python
LogReg = LogisticRegression()
DTree = DecisionTreeClassifier()
SVC = SVC(probability=True)
```

### Voting

```python
VotingHard = VotingClassifier(estimators=[('SVC', SVC), ('DTree', DTree), ('LogReg', LogReg)], voting='hard')
VotingHard.fit(X_train, y_train)
y_pred_votinghard = VotingHard.predict(X_test)

VotingSoft = VotingClassifier(estimators=[('SVC', SVC), ('DTree', DTree), ('LogReg', LogReg)], voting='soft')
VotingSoft.fit(X_train, y_train)
y_pred_votingsoft = VotingSoft.predict(X_test)
```

```python
from sklearn import metrics
metrics.accuracy_score(y_test, y_pred_votinghard)
```

```python
metrics.accuracy_score(y_test, y_pred_votingsoft)
```

### Bagging

```python
Bagging = BaggingClassifier()
Bagging.fit(X_train, y_train)
y_pred_bagging = Bagging.predict(X_test)
```

```python
metrics.accuracy_score(y_test, y_pred_bagging)
```

### Gradient Boosting Classifier

```python
GBC = GradientBoostingClassifier()
GBC.fit(X_train, y_train)
y_pred_gbc = GBC.predict(X_test)
```

```python
metrics.accuracy_score(y_test, y_pred_gbc)
```

### XGBoost

```python
XGB = XGBClassifier()
XGB.fit(X_train, y_train)
y_pred_xgb = XGB.predict(X_test)
```

```python
metrics.accuracy_score(y_test, y_pred_xgb)
```

### LightGBM

```python
lgbm = LGBMClassifier()
lgbm.fit(X_train, y_train)
y_pred_lgbm = lgbm.predict(X_test)
```

```python
metrics.accuracy_score(y_test, y_pred_lgbm)
```

### CatBoost

```python
catboost = CatBoostClassifier()
catboost.fit(X_train, y_train)
y_pred_cat = catboost.predict(X_test)
```

```python
metrics.accuracy_score(y_test, y_pred_cat)
```
