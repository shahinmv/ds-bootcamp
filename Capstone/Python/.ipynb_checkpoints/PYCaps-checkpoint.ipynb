```python
import pandas as pd
```

```python
df = pd.read_excel('data_cleaning.xlsx', sheet_name='Usage')
```

```python
df
```

```python
new_row = {df.columns[0]: df.columns[0]}
```

```python
df = df.append(new_row, ignore_index=True)
```

```python
df.columns[0]
```

```python
df.tail
```

```python
df[" 3 PM  Mon 24th-Mar-2014___0.384 kwh  "] = df[" 3 PM  Mon 24th-Mar-2014___0.384 kwh  "].str.lstrip("_ ")
df[" 3 PM  Mon 24th-Mar-2014___0.384 kwh  "] = df[" 3 PM  Mon 24th-Mar-2014___0.384 kwh  "].str.rstrip("kwh ")
```

```python
df = df.rename(columns={df.columns[0]:'cleaning'})
```

```python
df
```

```python
df['time'] = df['cleaning'].str.extract('(\d+)')
```

```python
df['cleaning'] = df['cleaning'].str.replace('^\d+', '')
df['cleaning'] = df['cleaning'].str.lstrip(" ")
```

```python
df
```

```python
df['ampm'] = df['cleaning'].str.extract('(AM|PM)')
```

```python
df['cleaning'] = df['cleaning'].str.replace('^(AM|PM)', '')
df['cleaning'] = df['cleaning'].str.lstrip(" ")
```

```python
df
```

```python
df['cleaning'] = df['cleaning'].str.replace('^\D+', '')
```

```python
df
```

```python
df['cleaning'] = df['cleaning'].str.replace('_', ' ')
```

```python
df
```

```python
df = df.assign(month=df['cleaning'].str.split().str[0],
               kwh=df['cleaning'].str.split().str[1])
```

```python
df = df.drop(columns=['cleaning'])
```

```python
df['month'] = df['month'].str.replace('(rd|nd|th|st)', '')
```

```python
df
```

```python
df['day'] = df["month"].str.split("[^0-9]").str.get(0)
```

```python
df['month'] = pd.to_datetime(df['month'])
```

```python
df['weekday'] = df['month'].dt.weekday
```

```python
df
```

```python
df['month'] = df['month'].dt.month
```

```python
df
```

```python
df
```

```python
df = df.sort_values(by=["month", "day", "ampm", "time"])
```

```python
df
```

```python
weekday_names = {
    0: 'Monday',
    1: 'Tuesday',
    2: 'Wednesday',
    3: 'Thursday',
    4: 'Friday',
    5: 'Saturday',
    6: 'Sunday'
}

# Use the map() method to map the numeric values of the weekday to their names
df['weekday'] = df['weekday'].map(weekday_names)
```

```python
month_names = {
    1: 'January',
    2: 'February',
    3: 'March',
    4: 'April',
    5: 'May',
    6: 'June',
    7: 'July',
    8: 'August',
    9: 'September',
    10: 'October',
    11: 'November',
    12: 'December'
}

# Use the map() method to map the numeric values of the month to their names
df['month'] = df['month'].map(month_names)
```
